```python
#cleaning

import pandas as pd
import numpy as np

df = pd.read_csv('/content/finaldata/june23ltimaccessible.csv', header=4)

clean = df.replace(',', '', regex=True)
clean.iloc[:, 2:] = clean.iloc[:, 2:].replace('-', np.nan, regex=True).astype(float)

clean.head()
```

```python
#sorting data
import statsmodels.api as sm
net_df = clean[clean['Flow'] == 'Net']

net_df.head()

net_df['Treatment'] = np.where(net_df['All Nationalities'] == 'EU', 1, 0)
net_df['Time'] = np.where(net_df['Period']>='YE Sep 16',1,0)
net_df['Treatment*Time'] = net_df['Treatment'] * net_df['Time']
model = ols(net_df['All Nationalities'], sm.add_constant(net_df[['Treatment', 'Time', 'Treatment*Time']])).fit()

```
