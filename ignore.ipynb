```python
#cleaning the data

import pandas as pd
import numpy as np

df = pd.read_csv('/content/finaldata/implicitandexplicitdata.csv')

clean = df.replace(',', '', regex=True)
clean.iloc[:, 1:] = clean.iloc[:, 1:].replace('-', np.nan, regex=True).astype(float)

df = pd.DataFrame(clean).reset_index()

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

clean.set_index(clean.columns[0], inplace=True)
germany_data = clean.loc['Germany']
plt.bar(germany_data.index, germany_data)
plt.xlabel('Languages')
plt.ylabel('Percentage')
plt.title(f'Percentage of Jobs Expicitly or Implictly Demanding Each Language in Germany')

plt.xticks(rotation=45, ha='right')
plt.ylim(0, 100)

plt.show()
```

```python
df = pd.read_csv('/content/finaldata/implicitandexplicitdata.csv')

clean = df.replace(',', '', regex=True)
clean.iloc[:, 1:] = clean.iloc[:, 1:].replace('-', np.nan, regex=True).astype(float)

clean.head()

import pandas as pd

clean.set_index('Country', inplace=True)
row_means = clean.mean(axis=1)
row_means_df = pd.DataFrame({'RowMeans': row_means})
print(row_means_df)
```

```python
row_means_df.plot.bar(rot=0, legend=False)
plt.xlabel('Countries')
plt.ylabel('Mean Value')
plt.title('Mean Value of Each Row')

plt.show()
```
